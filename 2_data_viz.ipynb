```
train = pd.read_pickle('train.p')
train.keys()
```

```
X_train, y_train = train['features'], train['labels']
X_train.shape, y_train.shape

#number of photos, pixels, pixels, number of channels - 3 beacause RGB
#each pixel within channel can have value between 0 to 255
```

```
#one pixel is represneted by 3 numbers (R, G, B), each picture is 32x32 pictures
plt.imshow(X_train[14224])
```

```
signs = pd.read_csv('signnames.csv')
signs
```

```
dict_signs = signs.to_dict()['b']
```

```
X_train[ y_train == 10].shape
```

```
#visualize data

for id_sign in dict_signs.keys():
  
  given_signs = X_train[ y_train == id_sign]
  plt.figure(figsize=(15,5))

  for i in range(9):
    plt.subplot('19{0}'.format(i+1))
    plt.imshow(given_signs[i])
    plt.axis('off')

plt.tight_layout
plt.show()
```

```
cnt = Counter(y_train).most_common()
id_labels, cnt_label = zip(*cnt)
ids = range(len(id_labels))

plt.figure(figsize=(15,5))
plt.bar(ids,cnt_label)

plt.xlabel('Sign')
labels =  [dict_signs[ id_labels[id_]] for id_ in id_labels]
plt.xticks(ids,labels, rotation = 'vertical')
plt.title('Road signs - appereneces by sign')
plt.show()
```
